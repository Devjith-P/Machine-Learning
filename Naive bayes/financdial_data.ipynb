```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for i in df.columns:
    if df[i].dtype == object:
        df[i] = le.fit_transform(df[i])
```

```python
x = df.iloc[:,:-2].values
```

```python
y_fraud = df.iloc[:,-2].values
y_flag_fraud = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train_fraud,y_test_fraud = train_test_split(x,y_fraud,test_size=.3,random_state=1)

x_train_flag,x_test_flag,y_train_flag,y_test_flag = train_test_split(x,y_flag_fraud,test_size=.3,random_state=1)
```

```python
from sklearn.svm import SVC
model_test = SVC()
model_test.fit(x_train,y_train_fraud)
```

```python
from sklearn.naive_bayes import MultinomialNB
model_fraud = MultinomialNB()
model_flag = MultinomialNB()
model_fraud.fit(x_train,y_train_fraud)
model_flag.fit(x_train,y_train_flag)
```

```python
y_pred_fraud = model_fraud.predict(x_test)
y_pred_flag = model_flag.predict(x_test_flag)
```

```python
y_test__ = model_test.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

print("Fraud",accuracy_score(y_test_fraud,y_pred_fraud)*100)
print("Flag",accuracy_score(y_test_flag,y_pred_flag)*100)
```

```python
print("Fraud\n",confusion_matrix(y_test_fraud,y_pred_fraud))
print("Flag\n",confusion_matrix(y_test_flag,y_pred_flag))
```
