```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import OneHotEncoder

one = make_column_transformer((OneHotEncoder(handle_unknown="ignore"),['State']),remainder='passthrough')
x = one.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
y_test
```

```python
df2 = pd.DataFrame(
    {
        "Actual": y_test,
        "Predicted":y_pred,
        "Difference":y_test-y_pred
    }
)
```

```python
df2
```

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, mean_squared_error,root_mean_squared_error,r2_score

mae = mean_absolute_error(y_test,y_pred)
mape = mean_absolute_percentage_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = root_mean_squared_error(y_test,y_pred)
r2score = r2_score(y_test,y_pred)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
print(f" Mean Absolute error {mae:0.2f},\n Mean absolute percentage error {mape:0.2f},\n Mean squared error {mse:0.2f},\n Root mean squared error {rmse:0.2f},\n r2 score {r2score:0.2f}")
```
