```python

df['12_gpa'] = df['12_gpa'].fillna(df['12_gpa'].mean())
df['last_year'] = df['last_year'].fillna(df['last_year'].mode()[0])
df['faculty'] = df['faculty'].fillna(df['faculty'].mode()[0])
df['accomodation'] = df['accomodation'].fillna(df['accomodation'].mode()[0])
df['monthly_allowance'] = df['monthly_allowance'].fillna(df['monthly_allowance'].mode()[0])
df['classes_missed'] = df['classes_missed'].fillna(df['classes_missed'].mode()[0])
df['romantic_relation'] = df['romantic_relation'].fillna(df['romantic_relation'].mode()[0])
df['parents_approve'] = df['parents_approve'].fillna(df['parents_approve'].mode()[0])
df['parents_relationship'] = df['parents_relationship'].fillna(df['parents_relationship'].mode()[0])
```

```python
df_enc = pd.get_dummies(df[['sex','last_year','faculty','accomodation','monthly_allowance','scholarship','additona_study_hrs','hours_partying','no_of_drinks','classes_missed','modules_failed','romantic_relation','parents_approve','parents_relationship']],drop_first=True,dtype=int)
```

```python
df = pd.concat([df,df_enc],axis=1)
```

```python
df = df.drop(columns=['sex','last_year','faculty','accomodation','monthly_allowance','scholarship','additona_study_hrs','hours_partying','no_of_drinks','classes_missed','modules_failed','romantic_relation','parents_approve','parents_relationship'])
```

```python
df
```

```python
x = df.drop(columns='2023_gpa')
y = df['2023_gpa']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
model = RandomForestRegressor(n_estimators=200,n_jobs=-1)

model.fit(x_train,y_train)
y_pred = model.predict(x_test)

# model = LinearRegression()
# model.fit(x_train,y_train)
# y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,root_mean_squared_error,r2_score

mae = mean_absolute_error(y_test,y_pred)
mape = mean_absolute_percentage_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = root_mean_squared_error(y_test,y_pred)
r2score = r2_score(y_test,y_pred)

print(f" Mean Absolute error {mae},\n Mean absolute percentage error {mape},\n Mean squared error {mse},\n Root mean squared error {rmse},\n r2 score {r2score}")
```
