```python
import pandas as pd
```

```python
df = pd.read_csv("/home/devjith/Luminar/Machine Learning/files/syntetic_linear_regression.csv")
```

```python
df
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=.3,random_state=42)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x,y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("X vs Y")
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_test,y_pred,color = 'r')
```

```python
print(model.coef_)
```

```python
print(model.intercept_)
```

```python
df2 = pd.DataFrame({
    "Actual value":y_test,
    "Predicted Value": y_pred,
    "Difference ": y_test-y_pred
})
```

```python
df2
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,root_mean_squared_error,r2_score

mae = mean_absolute_error(y_test,y_pred)
mape = mean_absolute_percentage_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = root_mean_squared_error(y_test,y_pred)
r2score = r2_score(y_test,y_pred)
```

```python
print(f" Mean Absolute error {mae},\n Mean absolute percentage error {mape},\n Mean squared error {mse},\n Root mean squared error {rmse},\n r2 score {r2score}")
```
