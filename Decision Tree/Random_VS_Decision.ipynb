```python
x = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

model_decision = DecisionTreeClassifier(criterion='entropy',random_state=42,max_depth=None,max_features=None)

model_regression = RandomForestClassifier(n_estimators=100,criterion='entropy',n_jobs=-1,random_state=42)

model_decision.fit(x_train,y_train)
model_regression.fit(x_train,y_train)
```

```python
y_pred_decision = model_decision.predict(x_test)
y_pred_reg = model_regression.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,ConfusionMatrixDisplay

print("Decision Tree")
print(classification_report(y_test,y_pred_decision))
cm = confusion_matrix(y_test,y_pred_decision)
print()
print("RandomForestClassifier")
print(classification_report(y_test,y_pred_reg))
cm2 = confusion_matrix(y_test,y_pred_decision)
```
