Simple Linear Regression
====

AREA        PRICE

2600        5500000

3000        6100000

3200        6250000

3600        6800000

3500            ?




Straight line equation Y = mX + c

Y = Output ,
m = Slope,
X = input,
C = constant or y intercept on X


The difference between predicted and actual value is called ERROR
--------------------------------------------------------------------------------------------------------------

```python
area = [2600,3000,3200,3600,4000]
price = [5500000,6100000,6250000,6800000,7100000]
```

```python
import matplotlib.pyplot as plt

plt.scatter(area,price)
```

    A straight line is drawn using the points

    From the data on x axis draw a perpendicular line to meet the straight line.

    After that draw a line perpendicular to y axis from the point on the straight line and the point where it meets the y axis is the prediction
