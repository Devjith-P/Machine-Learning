```python
df['Brand'] = df['Name'].apply(lambda x:x.split()[0])
```

```python
df_encoded = pd.get_dummies(df[['Location','Fuel_Type','Transmission','Owner_Type','Brand']],drop_first=True,dtype=int)
```

```python
df_encoded
```

```python
df = pd.concat([df,df_encoded],axis=1)
```

```python
df.columns.unique()
```

```python
df = df.drop(columns=['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','New_Price','Brand'])
```

```python
df.dtypes
```

```python
df['Mileage'] = df['Mileage'].str.replace("km/kg","")
df['Mileage'] = df['Mileage'].str.replace("kmpl","")
df['Engine'] = df['Engine'].str.replace("CC","")
df['Power'] = df['Power'].str.replace("bhp",'')
```

```python
df.isna().sum()
```

```python
df['Mileage'] = df['Mileage'].str.replace("null",'0')
df['Power'] = df['Power'].str.replace("null",'0')
df['Engine'] = df['Engine'].str.replace("null",'0')
```

```python
df.isna().sum()
```

```python
df['Engine'] = df['Engine'].astype(float)
df['Power'] = df['Power'].astype(float)
df['Mileage'] = df['Mileage'].astype(float)
```

```python
df.dtypes
```

```python
df.loc[df.Mileage==0,'Mileage']=np.nan
df.loc[df.Power==0,'Power']=np.nan
df.loc[df.Engine==0,'Engine']=np.nan
```

```python
df.isna().sum()
```

```python
df['Engine'].unique()
```

```python
df['Seats'].unique()
```

```python
df['Mileage'] = df['Mileage'].fillna(df['Mileage'].mean())
df['Power'] = df['Power'].fillna(df['Power'].mode()[0])
df['Engine'] = df['Engine'].fillna(df['Engine'].mean())
df['Seats'] = df['Seats'].fillna(df['Seats'].mode()[0])
```
